```python
import keras
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense,LSTM,Dropout
from keras.optimizers import  Adam,RMSprop
from keras.wrappers.scikit_learn import KerasRegressor
import pandas as pd
import numpy as np
import sklearn
from sklearn import metrics
from sklearn.model_selection import GridSearchCV
from keras import backend as K
from keras.callbacks import TensorBoard
config = tf.compat.v1.ConfigProto( device_count = {'GPU': 1} ) 
sess = tf.compat.v1.Session(config=config) 
tf.compat.v1.keras.backend.set_session(sess)

def create_model(lr,size,drop):
    rm=RMSprop(
    learning_rate=lr)
    shape=3
    window=20
    model = Sequential()
    model.add(LSTM(size,input_shape=(window,shape),return_sequences=True))
    model.add(Dropout(drop))
    model.add(LSTM(size,return_sequences=True))
    model.add(Dropout(drop))
    model.add(LSTM(size,return_sequences=False))
    model.add(Dropout(drop))
    model.add(Dense(shape))
    model.compile(loss='mae', optimizer=rm,metrics='cosine_similarity')
    return model

model = KerasRegressor(build_fn=create_model, epochs=20, batch_size=512, verbose=0)
batch_size = (500,100,1000)
lr=(0.005,0.003,0.001)
epochs = [30]
size=[500,300,100,50]
drop=[0.5,0.3,0.1]
param_grid = dict(batch_size=batch_size, epochs=epochs,lr=lr,drop=drop,size=size)
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=1,verbose=2,cv=2)
x_train, x_test, y_train, y_test = shifter(dataset,20)
grid_result = grid.fit(x_train, y_train,verbose=2)
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
for mean, stdev, param in zip(means, stds, params):
    print("%f (%f) with: %r" % (mean, stdev, param))
```
